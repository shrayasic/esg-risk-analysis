```python
import pandas as pd
import numpy as np
```

```python
!pip install openpyxl


# Load ESG Industrial Dataset
file_path = "../data/raw/esg_dataset.xlsx"

# Try to read the first sheet
df_esg = pd.read_excel(file_path)

# Show the first 5 rows
df_esg.head()
```

```python
df_esg.tail()
```

```python
# Dataset shape
print("Rows:", df_esg.shape[0])
print("Columns:", df_esg.shape[1])

# List column names
df_esg.columns.tolist()
```

```python
# Check data types
df_esg.dtypes

# Count missing values
df_esg.isnull().sum().sort_values(ascending=False)
```

```python
df_esg.columns = df_esg.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df_esg.head(2)
```

```python
df_esg['date'] = pd.to_datetime(df_esg['date'], errors='coerce')
df_esg['date'].head() 
```

```python
df_esg = df_esg.drop(columns=['water_recycle', 'turnover_empl'])
```

```python
# Drop rows with null in essential score columns
df_esg = df_esg.dropna(subset=['esg_score', 'social_score', 'gov_score', 'env_score'])

# Fill small missing values with median (example)
df_esg['pe_ratio'] = df_esg['pe_ratio'].fillna(df_esg['pe_ratio'].median())
```

```python
print("Rows:", df_esg.shape[0])
print("Columns:", df_esg.shape[1])
```
