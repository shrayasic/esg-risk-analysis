```python
import pandas as pd

file_path = "../data/raw/esg_dataset.xlsx"

df = pd.read_excel(file_path)  # Replace with the actual file name

# Show all column names
print("📌 Column Names:\n", df.columns.tolist())

# Basic info: datatypes, nulls
print("\n📊 Dataset Info:")
print(df.info())

# Quick preview of data
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Check % of missing values
missing_percent = df.isnull().mean().sort_values(ascending=False) * 100
print("📌 Columns with Missing Values (%):\n", missing_percent[missing_percent > 0])

# Visual heatmap of missing data
plt.figure(figsize=(15, 8))
sns.heatmap(df.isnull(), cbar=False, cmap='coolwarm', yticklabels=False)
plt.title("🔍 Missing Value Heatmap")
plt.show()
```

```python
# 1. Drop high-missing columns
df.drop(columns=[
    'Water_recycle', 'Turnover_empl', 'Women_Employees', 
    'Water_use', 'Injury_rate'  # optionally: 'Energy_use'
], inplace=True)

# 2. Impute low/mid-missing features
from sklearn.impute import SimpleImputer
import numpy as np

# Fill binary-style metrics with 0 (assuming 0 = no initiative/violation)
bin_cols = ['Strikes', 'Bribery', 'Recycling_Initiatives', 'Human_Rights']
df[bin_cols] = df[bin_cols].fillna(0)

# Impute ESG components with mean
for col in ['ESG_score', 'Social_score', 'Gov_score', 'Env_score']:
    df[col] = df[col].fillna(df[col].mean())

# Impute financial columns with mean
fin_cols = ['BVPS', 'Shares', 'Net_income', 'RETURN_ON_ASSET', 'QUICK_RATIO',
            'ASSET_GROWTH', 'FNCL_LVRG', 'PE_RATIO']
df[fin_cols] = df[fin_cols].fillna(df[fin_cols].mean())

# Median for board-related
df[['Board_Size', 'Board_gen_div', 'Shareholder_Rights']] = df[['Board_Size', 'Board_gen_div', 'Shareholder_Rights']].fillna(df[['Board_Size', 'Board_gen_div', 'Shareholder_Rights']].median())

# Example: Fill with industry median
df['CO2_emissions'] = df.groupby('Industry')['CO2_emissions'].transform(lambda x: x.fillna(x.median()))
df['Scope_1'] = df.groupby('Industry')['Scope_1'].transform(lambda x: x.fillna(x.median()))
df['Scope_2'] = df.groupby('Industry')['Scope_2'].transform(lambda x: x.fillna(x.median())) 
```
