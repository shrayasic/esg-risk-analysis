```python
def analyze_distribution(df):
    numeric_cols = ['total_esg_risk_score', 'environment_risk_score', 
                    'governance_risk_score', 'social_risk_score', 'controversy_score']

    for col in numeric_cols:
        sns.histplot(df[col], kde=True, color='skyblue')
        plt.axvline(df[col].mean(), color='red', linestyle='--', label=f"Mean: {df[col].mean():.2f}")
        plt.axvline(df[col].median(), color='green', linestyle='--', label=f"Median: {df[col].median():.2f}")
        plt.title(f'Distribution of {col} | Skew: {df[col].skew():.2f}')
        plt.legend()
        plt.show()


def detect_outliers_iqr(df):
    numeric_cols = ['total_esg_risk_score', 'environment_risk_score', 
                    'governance_risk_score', 'social_risk_score']
    
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR
        outliers = df[(df[col] < lower) | (df[col] > upper)]
        
        print(f"{col}: {len(outliers)} outliers")
        sns.boxplot(x=df[col], color='lightgreen')
        plt.title(f'Boxplot of {col}')
        plt.show()


def esg_heatmap_by_industry(df):
    # Top 10 industries to keep it readable
    top_industries = df['industry'].value_counts().head(10).index
    filtered_df = df[df['industry'].isin(top_industries)]

    # Group and calculate average scores
    industry_group = filtered_df.groupby('industry')[['environment_risk_score', 
                                                       'governance_risk_score', 
                                                       'social_risk_score']].mean()
    
    sns.heatmap(industry_group, annot=True, cmap='YlOrRd', fmt='.1f')
    plt.title("Average ESG Component Scores by Industry (Top 10)")
    plt.xlabel("ESG Components")
    plt.ylabel("Industry")
    plt.show()


def statistical_summary(df):
    numeric_cols = ['total_esg_risk_score', 'environment_risk_score', 
                    'governance_risk_score', 'social_risk_score', 'controversy_score']

    print("=== Descriptive Statistics ===")
    print(df[numeric_cols].describe().round(2))

    print("\n=== Correlation Matrix (Pearson) ===")
    corr = df[numeric_cols].corr()
    print(corr.round(2))

    print("\n=== Risk Level Distribution ===")
    print(df['esg_risk_level'].value_counts())

    print("\n=== Sector-wise ESG Risk Summary ===")
    print(df.groupby('sector')['total_esg_risk_score'].agg(['count', 'mean', 'std']).round(2).sort_values('mean', ascending=False))


def run_balanced_eda(df):
    print("\n1. Distribution & Skewness")
    analyze_distribution(df)

    print("\n2. Outlier Analysis")
    detect_outliers_iqr(df)

    print("\n3. ESG Component Heatmap by Industry")
    esg_heatmap_by_industry(df)

    print("\n4. Statistical Summary")
    statistical_summary(df)


run_balanced_eda(df)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns

# Selecting numerical features
num_cols = ['total_esg_risk_score', 'environment_risk_score', 
            'social_risk_score', 'governance_risk_score', 'controversy_score']

X = df[num_cols].dropna()  # Drop rows with missing values

# Standardize
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA()
pca_components = pca.fit_transform(X_scaled)

# Explained variance
explained = pca.explained_variance_ratio_

# Scree plot
plt.figure(figsize=(8,5))
sns.lineplot(x=range(1, len(explained)+1), y=explained, marker='o')
plt.title('Scree Plot - PCA')
plt.xlabel('Principal Component')
plt.ylabel('Variance Explained')
plt.grid(True)
plt.show()
```

```python
!pip install statsmodels
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd

# VIF Calculation
vif_data = pd.DataFrame()
vif_data['feature'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X_scaled, i) for i in range(X_scaled.shape[1])]

print(vif_data)
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Use first 2 PCA components for clustering
X_pca = PCA(n_components=2).fit_transform(X_scaled)

# Try different values of K
sil_scores = []
for k in range(2, 7):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X_pca)
    sil_scores.append(silhouette_score(X_pca, labels))

# Plot Silhouette Scores
plt.figure(figsize=(6,4))
sns.lineplot(x=range(2, 7), y=sil_scores, marker='o')
plt.title('Silhouette Score vs Number of Clusters')
plt.xlabel('k')
plt.ylabel('Silhouette Score')
plt.grid(True)
plt.show()

# Final Clustering with best K (say k=3)
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_pca)

# Add to dataframe
df['esg_cluster'] = clusters

# Visualize
plt.figure(figsize=(7,5))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=clusters, palette='viridis')
plt.title('KMeans Clustering (PCA Reduced Data)')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend(title='Cluster')
plt.show()
```

```python
def create_outlier_flag(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    return ((df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))).astype(int)

df['env_outlier'] = create_outlier_flag('environment_risk_score')
df['gov_outlier'] = create_outlier_flag('governance_risk_score')
df['soc_outlier'] = create_outlier_flag('social_risk_score')
```
